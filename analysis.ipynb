## Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Encode Categorical Features including the Target
label_encoder = LabelEncoder()
df_encoded = df.copy()

for column in categorical_features:
    df_encoded[column] = label_encoder.fit_transform(df_encoded[column])

df_encoded['Target_encoded'] = label_encoder.fit_transform(df_encoded['Target'])

# Step 2: Select only numerical columns including the encoded target
numerical_columns = numerical_features + binary_features + ['Target_encoded']

# Step 3: Calculate Correlation Matrix
correlation_matrix = df_encoded[numerical_columns].corr()

# Step 4: Extract Correlations with Target
target_correlations = correlation_matrix['Target_encoded'].sort_values(ascending=False)

# Display the correlations with the Target
print("Correlation of features with Target:")
print(target_correlations)

# Step 5: Visualize Correlation using a Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Step 6: Visualize Correlations with Target using a Bar Plot
plt.figure(figsize=(12, 6))
target_correlations.drop('Target_encoded').plot(kind='bar')
plt.title('Correlation of Features with Target')
plt.xlabel('Features')
plt.ylabel('Correlation Coefficient')
plt.show()
```

We see that Unemployment rate, International, Educational special needs have really low close to 0 correlation with Target. We can safely exclude them. 

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

# Load Training Data
datafile_train_path = 'data/train.csv'
df_train = pd.read_csv(datafile_train_path)

# Load Test Data
datafile_test_path = 'data/test.csv'
df_test = pd.read_csv(datafile_test_path)

categorical_features = ['Marital status', 'Application mode', 'Application order', 'Course',
                        'Previous qualification', "Mother's qualification", "Father's qualification",
                        "Mother's occupation", "Father's occupation"]

numerical_features = ['Previous qualification (grade)', 'Age at enrollment',
                      'Curricular units 1st sem (enrolled)', 'Curricular units 1st sem (evaluations)',
                      'Curricular units 1st sem (approved)', 'Curricular units 1st sem (grade)',
                      'Curricular units 2nd sem (enrolled)', 'Curricular units 2nd sem (evaluations)',
                      'Curricular units 2nd sem (approved)', 'Curricular units 2nd sem (grade)',
                      'GDP', 'Admission grade', 'Inflation rate']

binary_features = ['Daytime/evening attendance', 'Displaced',
                   'Debtor', 'Tuition fees up to date', 'Scholarship holder', 'Gender']

target_feature = 'Target'

# Encode Categorical Features
label_encoder = LabelEncoder()

# Handle unseen labels by fitting the encoder on the combined data
combined_data = pd.concat([df_train, df_test], ignore_index=True)
for column in categorical_features:
    label_encoder.fit(combined_data[column])
    df_train[column] = label_encoder.transform(df_train[column])
    df_test[column] = label_encoder.transform(df_test[column])

# Encode Target Feature
df_train['Target_encoded'] = label_encoder.fit_transform(df_train['Target'])

# Prepare Training Data
X_train = df_train[numerical_features + categorical_features + binary_features]
y_train = df_train['Target_encoded']

# Prepare Test Data
X_test = df_test[numerical_features + categorical_features + binary_features]
y_test = df_test['Target']  # actual target labels for evaluation

# Standardize Numerical Features
scaler = StandardScaler()
X_train[numerical_features] = scaler.fit_transform(X_train[numerical_features])
X_test[numerical_features] = scaler.transform(X_test[numerical_features])

# Check Data Shapes
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")

# Define Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier()
}

# Train and Evaluate Models
results = {}

for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on Test Data
    y_pred = model.predict(X_test)
    
    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)
    confusion = confusion_matrix(y_test, y_pred)
    
    # Store results
    results[model_name] = {
        "accuracy": accuracy,
        "f1_score": f1,
        "classification_report": report,
        "confusion_matrix": confusion
    }
    
    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy}")
    print(f"F1 Score: {f1}")
    print("Classification Report:")
    print(report)
    print("Confusion Matrix:")
    print(confusion)
    print("\n" + "="*60 + "\n")
```
